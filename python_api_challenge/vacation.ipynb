```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
import time

# Import API key
from api_keys import g_key
```

```python
#Load part 1 data from csv

weather_data = pd.read_csv("cities.csv")
weather_data
```

```python
#Create a heat map that displays the humidity for every city from Part I.

gmaps.configure(api_key=g_key)
# Store 'Lat' and 'Lng' into  locations 
locations = weather_data[["Lat", "Lng"]].astype(float)

humidity = weather_data["Humidity"]
```

```python
# Plot Heatmap
fig = gmaps.figure(center=(46.0, -5.0), zoom_level=2)
max_intensity = np.max(humidity)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights = humidity, dissipating=False, max_intensity=100, point_radius=3)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
## Narrow down the cities with wind speed less than 10 mph, cloudiness equals to 0 and max temp between 60 and 80
new_city_df = weather_data.loc[(weather_data["Wind Speed"]<=10) & (weather_data["Cloudiness"] == 0)&(weather_data["Max Temp"]>=60) & (weather_data["Max Temp"]<=80)].dropna()
new_city_df
```

```python
# Store into variable named hotel_df.
# Add a "Hotel Name" column to the DataFrame.
# Set parameters to search for hotels with 5000 meters.
# Hit the Google Places API for each city's coordinates.
# Store the first Hotel result into the DataFrame.
# Plot markers on top of the heatmap

hotel_df = new_city_df.loc[:,["City","Country","Lat","Lng"]]
#Add hotel name to the dataframe
hotel_df["Hotel Name"] = ""
#Display Result
hotel_df
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {"type" : "hotel",
          "keyword" : "hotel",
          "radius" : 5000,
          "key" : g_key}



for index,row in hotel_df.iterrows():
    #get lag,lng,city
    lat = row["Lat"]
    lng = row["Lng"]
    cityname = row["City"]
    
    # add keyword to params dict
    params['location'] = f"{lat},{lng}"

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {cityname}.")
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        print(f"Closest hotel in {cityname}  is {results[0]['name']}.")
        hotel_df.loc[index, "Hotel Name"] = results[0]['name']
        
       
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
```

```python

# Display the hotel dataframe
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
 #Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content = hotel_info)
# Add the layer to the map
fig.add_layer(markers)

# Display Map
fig
```
